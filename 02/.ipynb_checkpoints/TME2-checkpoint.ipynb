````python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
````

````python
def MSE(y,val):
    '''y et val de même dimension et aligné sur les mêmes x'''
    return np.mean((y-val)*(y-val))
````

````python
data = np.loadtxt("./bluegills.txt", dtype=int, skiprows = 1)
````

````python
np.random.shuffle(data)
train = data[0:int(0.8*len(data))]
test = data[int(0.8*len(data)):]
````

````python
p = np.polyfit(train[:,0],train[:,1],0)
````

````python
x = np.unique(data[:,0])
plt.scatter(data[:,0], data[:,1])
plt.plot(x, np.polyval(p,x))
plt.ylabel("length")
plt.xlabel("age")
plt.show()
````

````python
x = np.unique(test[:,0])
plt.scatter(test[:,0], test[:,1])
plt.plot(x, np.polyval(p,x))
plt.ylabel("length")
plt.xlabel("age")
plt.show()
print("Score : " + str(MSE(test[:,1],np.polyval(p,test[:,0]))))
````

````python
def showTrain(data, deg):
    '''deg est une liste des degrées du polynome'''
    np.random.shuffle(data)
    train = data[0:int(0.8*len(data))]
    test = data[int(0.8*len(data)):]
    x = np.unique(test[:,0])
    plt.scatter(test[:,0], test[:,1])
    #p = np.apply_along_axis(lambda x : np.polyfit(train[:,0],train[:,1],x), 0, deg)
    pol = []
    for i in deg:
        pol.append(np.polyfit(train[:,0],train[:,1],i))
        plt.plot(x, np.polyval(pol[-1],x))
    plt.ylabel("length")
    plt.xlabel("age")
    plt.show()
    print("Valeurs du MSE :")
    m = []
    for i,p in zip(deg,pol):
        m.append(MSE(test[:,1],np.polyval(p,test[:,0])))
        print("Polynôme de degrée " + str(i) + " : " + str(m[-1]))
    ma = min(zip(deg,m), key = lambda x : x[1])[0]
    print("Degré du meilleur polynome : " + str(ma))
````

```showTrain``` mélange les données sous forme d'un ensemble de train et de test, génére un série de polynome sur les données de train et affiche les courbes et les scores de ces polynomes sur l'ensemble de test.

````python
data = np.loadtxt("./bluegills.txt", dtype=int, skiprows = 1)
np.random.seed(121344)
for i in range(5):
    showTrain(data, range(6))
````

On peut voir que le degrée du meilleur polynome dépends du découpage des données, on ne peut donc pas avoir une réponse stable avec ce genre de méthode.

````python
def showKFold(data, deg, k = 3, show = False):
    '''deg est une liste des degrées du polynome'''
    if  k < 1 :
        print("k < 1")
        return
    np.random.shuffle(data)
    fold = np.array_split(data, k)
    
    mse = []
    for i in range(k):
        if show :
            print("Kfold " + str(i))
        test = fold[i]
        train = np.concatenate(np.delete(fold, i))
        x = np.unique(test[:,0])
        pol = []
        for i in deg:
            pol.append(np.polyfit(train[:,0],train[:,1],i))
            if show :
                plt.plot(x, np.polyval(pol[-1],x))
        if show :
            plt.scatter(test[:,0], test[:,1])
            plt.ylabel("length")
            plt.xlabel("age")
            plt.show()
            print("Valeurs du MSE :")
        m = []
        for i,p in zip(deg,pol):
            m.append(MSE(test[:,1],np.polyval(p,test[:,0])))
            if show :
                print("Polynôme de degrée " + str(i) + " : " + str(m[-1]))
        mse.append(m)

    mse = np.block(mse)
    m = np.mean(mse, 0)
    
    print("\nRésultats :")
    for i,p in zip(deg,m):
        print("Polynôme de degrée " + str(i) + ", MSE moyen : " + str(p))
    ma = min(zip(deg,m), key = lambda x : x[1])[0]
    print("Degré du meilleur polynome : " + str(ma))
````

````python
data = np.loadtxt("./bluegills.txt", dtype=int, skiprows = 1)
np.random.seed(215846)
for i in range(5):
    showKFold(data, range(6), k = 5)
````

On peut voir que le résultat est plus stable mais pas parfaitement (ça se joue à peu pour 4è cas).

````python
def showLOO(data, deg):
    '''deg est une liste des degrées du polynome'''
    np.random.shuffle(data)
    
    mse = []
    for i in range(data.shape[0]):
        train = np.delete(data, i, 0)
        pol = []
        for i in deg:
            pol.append(np.polyfit(train[:,0],train[:,1],i))
        m = []
        for i,p in zip(deg,pol):
            m.append(MSE(data[i,1],np.polyval(p,data[i,0])))
        mse.append(m)
    
    mse = np.block(mse)
    m = np.mean(mse, 0)
    
    print("\nRésultats :")
    for i,p in zip(deg,m):
        print("Polynôme de degrée " + str(i) + ", MSE moyen : " + str(p))
    ma = min(zip(deg,m), key = lambda x : x[1])[0]
    print("Degré du meilleur polynome : " + str(ma))
````

````python
data = np.loadtxt("./bluegills.txt", dtype=int, skiprows = 1)
np.random.seed(587459)
for i in range(5):
    showLOO(data, range(6))
````

Ces résultats ne sont pas attendus, la méthode leave one out n'est pas sensé donner des résultats différents suivant le mélange des données, la fonction ```np.polyfit``` ayant l'air stable comme on peut le voir si dessous, le problème doit venir du code.

````python
def negLogVraissemblance(data, p):
    mse = MSE(data[:,1],np.polyval(p,data[:,0]))
    return data.shape[0] / 2 * np.log(mse*2*np.pi) + data.shape[0] / 2
````

````python
def AIC(k, l):
    return 2* k - 2 * l
````

````python
def AICShow(data, deg) :
    x = np.unique(data[:,0])
    plt.scatter(data[:,0], data[:,1])
    pol = []
    for i in deg:
        pol.append(np.polyfit(data[:,0],data[:,1],i))
        plt.plot(x, np.polyval(pol[-1],x))
    plt.ylabel("length")
    plt.xlabel("age")
    plt.show()
    print("Valeurs AIC :")
    m = []
    for i,p in zip(deg,pol):
        m.append(AIC(i+1, -negLogVraissemblance(data, p)))
        print("Polynôme de degrée " + str(i) + ", AIC : " + str(m[-1]))
    ma = min(zip(deg,m), key = lambda x : x[1])[0]
    print("Degré du meilleur polynome : " + str(ma))
````

````python
data = np.loadtxt("./bluegills.txt", dtype=int, skiprows = 1)
AICShow(data, range(6))
````

````python
def BIC(k, n, l):
    return -2 * l + k * np.log(n)
````

````python
def BICShow(data, deg) :
    x = np.unique(data[:,0])
    plt.scatter(data[:,0], data[:,1])
    pol = []
    for i in deg:
        pol.append(np.polyfit(data[:,0],data[:,1],i))
        plt.plot(x, np.polyval(pol[-1],x))
    plt.ylabel("length")
    plt.xlabel("age")
    plt.show()
    print("Valeurs BIC :")
    m = []
    for i,p in zip(deg,pol):
        m.append(BIC(i+1, data.shape[0], -negLogVraissemblance(data, p)))
        print("Polynôme de degrée " + str(i) + ", BIC : " + str(m[-1]))
    ma = min(zip(deg,m), key = lambda x : x[1])[0]
    print("Degré du meilleur polynome : " + str(ma))
````

````python
data = np.loadtxt("./bluegills.txt", dtype=int, skiprows = 1)
BICShow(data, range(6))
````

On peut voir que les différentes méthodes donnent des résultats plutôt convergeant (suivant la stabilité de la méthode), le polynome de degrée 2 semble le meilleur modèle.

````python
np.random.shuffle(data)
data = data[:int(data.shape[0]*0.8)]
deg = range(6)
showTrain(data, deg)
showKFold(data, deg)
showLOO(data, deg)
AICShow(data, deg)
BICShow(data, deg)
````

Le résultat à l'air plutôt stable si on diminue la taille des données, même si cela peut potentielement donner de l'instabilité.

On peut conclure que le modèle générant les vrais données semble être un polynome de degrée 2 bruité.
Suivant les données le vrais modèle peut être intéressant à trouver (ici par exemple, il peut donner des informations sur les mécanismes de croissance de ces poissons), mais sur des problèmes plus complexe, le modèle réel peut être plus difficile à approcher et à comprendre, il est alors moins important à trouver.

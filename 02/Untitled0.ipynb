```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
def MSE(y,val):
    '''y et val de même dimension et aligné sur les mêmes x'''
    return np.mean((y-val)*(y-val))
```

```python
data = np.loadtxt("./bluegills.txt", dtype=int, skiprows = 1)
```

```python
plt.scatter(data[:,0], data[:,1])
plt.show()
```

```python
np.random.shuffle(data)
train = data[0:int(0.8*len(data))]
test = data[int(0.8*len(data)):]
```

```python
p = np.polyfit(train[:,0],train[:,1],0)
```

```python
x = np.unique(data[:,0])
plt.scatter(data[:,0], data[:,1])
plt.plot(x, np.polyval(p,x))
plt.show()
```

```python
x = np.unique(test[:,0])
plt.scatter(test[:,0], test[:,1])
plt.plot(x, np.polyval(p,x))
plt.show()
print(MSE(test[:,1],np.polyval(p,test[:,0])))
```
